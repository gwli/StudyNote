```python
from __future__ import  division
from scipy.stats import bernoulli
import numpy as np
init_printing()
```

```python
p_true =1/2
fp = bernoulli(p_true)
def sample(n=10):
    return fp.rvs(n)
xs = sample(100)
```

```python
from sympy import  *
from sympy.abc import  x,z
p= symbols('p',positive =True)
```

```python
L = p**x*(1-p)**(1-x)
J = np.prod(L.subs(x,i) or i in xs)
J
```

为什么要使用替代那函数那。

```python
logJ = expand_log(log(J))
logJ
```

```python
sol = solve(diff(logJ,p),p)
sol
```

```python
x = linspace(0,1,100)
plot(x,map(lambdify(p,logJ,'numpy'),x),sol,logJ.subs(p,sol),'*',
     p_true,logJ.subs(p,p_true),'s',
     )
```
